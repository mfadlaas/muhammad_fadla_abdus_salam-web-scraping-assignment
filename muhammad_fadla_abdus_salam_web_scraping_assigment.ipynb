```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

List of countries by air pollution: 2020 list (UChicago ACLI 2020) - Wikipedia


All data are valid for the year 2020 and are taken from the Air Quality Life Index (ACLI) of the University of Chicago. In addition to particulate matter pollution, the modeled potential loss of life expectancy of the population due to particulate matter pollution is given

```python
url3 = 'https://en.wikipedia.org/wiki/List_of_countries_by_air_pollution#2018-2023_List_(IQAir_World_Air_Quality_Ranking)'
```

```python
r= requests.get(url3)
print(r)
```

```python
data  = requests.get(url3).text
soup1 = BeautifulSoup(data,"html.parser")
```

```python
tables = soup1.find_all('table')

# Finding the number of tables in the particular html content

len(tables)
```

```python
# Scrapping the first table from the page so go ahead with it
table1 = tables[0]

# Moving into all the rows in the selected table
body = table1.find_all("tr")

# Head values (Column names) are the first items of the body list so finding the 0th item
head = body[0]
body_rows = body[1:]
```

```python
# Creating an empty list to keep column names
headings = []
# Loop through all th elements
for item in head.find_all("th"):
    # Convert the th elements to text and strip "\n"
    item = (item.text).rstrip("\n")
    # Append the clean column name to headings
    headings.append(item)
# Print the Column names added into the empty list
print(headings)
```

```python
titles = []
for i in headings:
  title = i
  titles.append(title)

#print(titles)

df = pd.DataFrame(columns = titles)
print(df)
```

```python
for i in body[1:]:
  data = i.find_all("td")
  #print(data)
  row = [tr.text.strip() for tr in data]
  #print(row)
  l = len(df)
  df.loc[l] = row

#print(df)
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
# saving the dataframe into a csv file
df.to_csv('2020 list (UChicago ACLI 2020).csv')
```
